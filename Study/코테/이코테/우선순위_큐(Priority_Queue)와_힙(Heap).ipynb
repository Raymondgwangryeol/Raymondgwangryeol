## **우선순위 큐(Priority Queue)**
우선순위가 가장 높은 데이터를 가장 먼저 삭제하는 자료구조.    
데이터를 우선 순위에 따라 처리하고 싶을 때 사용.     
- 예) 물건 데이터를 자료구조에 넣었다가 가치가 높은 물건부터 꺼내서 확인해야 하는 경우      

<left>

|자료구조|먼저 추출되는 데이터|
|:-------|:-------------------|
|스택|가장 나중에 삽입된 데이터|
|큐|가장 먼저 삽입된 데이터|
|우선순위 큐|가장 우성순위가 높은 데이터|

</left>

<br><br>

### **우선순위 큐 구현 방법**
1. 리스트로
2. heap을 사용해서

<left>

|우선순위 큐 구현 방식|삽입 시간|삭제 시간|
|:-------------|:-----|:-----|
|리스트|O(1)|O(N)|
|힙|O(NlogN)|O(NlogN)|

</left>

힙을 사용하면 삽입, 삭제 시간 모두 최악의 경우에도 시간복잡도 **O(NlogN)**을 유지

<br><br>

### **힙(Heap)의 특징**
**완전 이진 트리** 자료 구조의 일종으로, 항상 루트 노드를 제거함.      
최소 힙 구성 함수는 **heapify()**.
- **최소 힙(min heap)**
    - 루트 노드가 가장 작은 값
    - 자식 노드들에 비해 부모 노드가 항상 작은 값을 가짐
    - 값이 작은 데이터가 우선적으로 제거 (오름차순)
- **최대 힙(max heap)**
    - 루트 노드가 가장 큰 값
    - 자식 노드들에 비해 부모 노드가 항상 큰 값을 가짐.
    - 값이 큰 데이터가 우선적으로 제거 (내림차순)
<br><br>

#### **완전 이진 트리**
루트 노드부터 시작해서 왼족 자식 노드, 오른쪽 자식 노드 순으로 데이터가 차례대로 삽입되는 트리.

#### **최소 힙에 원소 삽입**
1. 완전 이진 트리의 특징에 따라 데이터가 차례대로 삽입됨.
2. 삽입된 데이터와 부모 노드(루트 노드)를 비교하여, 삽입된 데이터가 값이 더 작다면 위치를 바꿈.
3. 위치를 바꿔가며 **루트 방향으로 상향식**으로 이동.
4. 최소 힙의 형태가 될 때까지 반복.

#### **최소 힙 원소 삭제**
1. 루트 노드가 삭제됨.
2. 맨 아래의 끝 노드를 루트 노드로 올림.
3. 아래의 자식 노드와 비교해서 루트 노드가 더 크면 자식노드와 자리를 바꿈.
4. 위치를 바꿔가며 **가장 아래쪽까지 하향식**으로 이동.
5. 최소 힙의 형태가 될 때 까지 반복.

#### **최소 힙 정렬 구현 예제**

```python
import heapq

def heapsort(iterable):
    h = []
    result = []
    # 모든 원소를 차례대로 힙에 삽입
    for value in iterable:
        heapq.heappush(h, value)
    # 모든 원소를 힙에서 차례로 꺼내 담기
    for i in range(len(h)):
        result.append(heapq.heappop(h))
    return result

n = int(input())
arr = []

for i in range(n):
    arr.append(int(input()))

res = heapsort(arr)

for i in res:
    print(i)
```
